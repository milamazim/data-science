```python
import pandas as pd
```

```python
clientes = '/content/drive/MyDrive/Colab Notebooks/PÓS_XP/Modulo_2_Big_data_Mysql/Desafio/dados_clientes_financeira.csv'
estado = '/content/drive/MyDrive/Colab Notebooks/PÓS_XP/Modulo_2_Big_data_Mysql/Desafio/estado_regiao.csv'
investimento = '/content/drive/MyDrive/Colab Notebooks/PÓS_XP/Modulo_2_Big_data_Mysql/Desafio/investimentos_clientes.csv'
```

```python
#importando os dados
dados_clientes = pd.read_csv(clientes, encoding= 'latin1',sep = ';')
dados_estados = pd.read_csv(estado, encoding= 'latin1', sep = ';')
dados_investimentos = pd.read_csv(investimento, encoding= 'latin1', sep = '|')
```

```python
dados_clientes = dados_clientes.drop_duplicates()
```

```python
dados_clientes.isnull().sum()
```

```python
dados_cliente_feminino = dados_clientes[dados_clientes['genero'] == 'Feminino']
dados_cliente_masculino = dados_clientes[dados_clientes['genero'] == 'Masculino']
```

```python
dados_cliente_masculino.head(3)
```

```python
moda_M = dados_cliente_masculino['escolaridade'].mode()
moda_M[0]
```

```python
moda_F = dados_cliente_feminino['escolaridade'].mode()
moda_F[0]
```

```python
dados_clientes['escolaridade'].fillna(moda_M[0], inplace= True)
```

```python
dados_clientes.isnull().sum()
```

```python
media_M = dados_cliente_masculino['idade'].mean().round(2)
media_M
```

```python
media_F = dados_cliente_feminino['idade'].mean().round(2)
media_F
```

```python
dados_clientes.loc[(dados_clientes['genero'] == 'Feminino') & (dados_clientes['idade'].isnull()), 'idade'] = media_F
```

```python
dados_clientes.loc[(dados_clientes['genero'] == 'Masculino') & (dados_clientes['idade'].isnull()), 'idade'] = media_M
```
